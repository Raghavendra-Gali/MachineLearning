```
import io
import pandas as pd

filename= "train_images"
# filename2= ""
data = pd.read_csv(io.StringIO(uploaded[filename].decode('utf-8')))
# images,lables = load_data(io.StringIO(uploaded[filename1].decode('utf-8')),io.StringIO(uploaded[filename2].decode('utf-8')))
```

```
data.head()
labels = data['Label']
data=data.drop(['Label'],axis=1)
```

```
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
```

```
temp_1000 = data[:1000]
```

```
# temp_1000.value_counts()
```

<h3>Visualizing first 1000 images in TSNE</h3>

```
temp_1k = TSNE(n_components=2,perplexity=45).fit_transform(temp_1000)
```

```
d={"Major":temp_1k[:,0],"Minor":temp_1k[:,1],"Labels":labels[:1000]}
z = pd.DataFrame(data=d)
import seaborn as sns
sns.FacetGrid(z,hue="Labels",size=6).map(plt.scatter,"Major","Minor").add_legend()
plt.show()
```
